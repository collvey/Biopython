## View Sample Data

```python
def read_data(filepath):
  with open(filepath, 'r') as f:
    data = f.readlines()
  return data

def view_data(text_data):
  output = []
  seq_format = []
  seq_anno = []
  seq = []
  for line in text_data:
    split_line = line.split()
    # print(split_line)
    if len(split_line) > 1:
      word = split_line[0]
      anno = split_line[1]
      seq.append(word)
      seq_anno.append(anno)
      seq_max_len = max(len(word), len(anno))
      seq_format.append(f'%-{seq_max_len+1}s')
    else:
      output.append(''.join(seq_format) % tuple(seq_anno))
      output.append(''.join(seq_format) % tuple(seq))
      output.append('')
      seq_format = []
      seq_anno = []
      seq = []
  return output
```

```python
print('\n'.join(view_data(read_data('./sample_data/train.bmes'))))
```

```python
print('\n'.join(view_data(read_data('./sample_data/train.cappos.bmes'))))
```
